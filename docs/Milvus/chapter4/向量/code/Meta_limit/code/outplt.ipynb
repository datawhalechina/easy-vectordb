```python
import json
import matplotlib.pyplot as plt

with open('result_d4_k2.json') as f:
    results = json.load(f)

N_list = [r["parameters"]["n"] for r in results]
Acc_list = [r["metrics"]["final_accuracy"] for r in results]

plt.plot(N_list, Acc_list, marker='o')
plt.xlabel("N")
plt.ylabel("final_accuracy")
plt.title("Critical N Search")
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

#  d 
all_results = {}

#  d=4 
with open('result_d4_k2.json') as f:
    all_results['d=4'] = json.load(f)

#  d=5 
with open('result_d5_k2.json') as f:
    all_results['d=5'] = json.load(f)

#  d=6 
with open('result_d6_k2.json') as f:
    all_results['d=6'] = json.load(f)

plt.figure(figsize=(10, 6)) # 

for d_label, results in all_results.items():
    N_list = [r["parameters"]["n"] for r in results]
    Acc_list = [r["metrics"]["final_accuracy"] for r in results]
    plt.plot(N_list, Acc_list, marker='o', label=d_label)

plt.xlabel("N (Number of Documents)")
plt.ylabel("Final Accuracy")
plt.title("Critical N Search for Different Embedding Dimensions (k=2)")
plt.legend() #  d 
plt.grid(True) # 
plt.show()
```
